### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols

# build the formula 
f = "sales~TV"
# create a fitted model in one line
model = ols(formula=f, data=data).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"


Note here that the coefficients represent associations, not causations


```python
# the R-squared is 0.61, meaning this is the amount of variation between sales
#and TV budget changes that can be explained by the model
#the coefficient intercept explains that a unit increase in TV spending
#is associated with a 0.0475 increase in sales
```

### Step 6:  Draw a prediction line with data points omn a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
tv_df = pd.DataFrame({'TV': [data.TV.min(), data.TV.max()]})
print(tv_df.head())
# make predictions for those x values and store them
predict = model.predict(tv_df)
print(predict)

# first, plot the observed data and the least squares line
data.plot(kind='scatter', x='TV', y='sales')
plt.plot(tv_df, predict, c="red", label="model predict")
plt.legend()
plt.show()
```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

```python
# Record Your observations on residuals
#graph 1 - there is a positive correlation within the confidence interval,
#however, the second graph shows a cone-shape emerging, suggesting that the
#error terms are not homoscedastic. the model's errors versus the specified
#predictor variable (TV) shows signs of heteroskedasticity, breaking one of
#the assumptions of OLS
```

### Next, repeat above for radio and go through the same process, recording your observations

```python
#Creating the regression formula
f = "sales~radio"
model=ols(formula=f, data=data).fit()

#summary of the model
model.summary()
```

```python
#The R-sqared is not as high, and only 33% of the variation in sales can be
#explained by radio ad budgeting amounts. 
```

```python
#creating a dataframe with min and max of radio
radio_df = pd.DataFrame({'radio': [data.radio.min(), data.radio.max()]})
print(radio_df.head())

#making some predictions for the radio x values:
preds = model.predict(radio_df)
print(preds)
```

```python
#the goods:
print("R-Squared: ", model.rsquared)
print(model.params)

data.plot(kind="scatter", x="radio", y="sales")
plt.plot(radio_df, preds, c="red")
plt.show()

fig=plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "radio",fig=fig)
plt.show()
```

```python
model.summary()
```

```python
# Record your observations here for goodnes of fit 
#radio is a weaker predictor than TV, and also shows signs of heteroskedasticity
#the R-squared is pretty weak at .33, and the it has a higher amount 
#of skewness and kurtosis than tv.
```
